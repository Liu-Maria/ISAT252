## import data frame plot

```python
import pandas
```

```python
df = pandas.read_excel('s3://isat252-liu/house_price.xls')
df[:10]
```

```python
df.plot(x = 'built_in', y = 'price')
```

## line chart
average prices per year

```python
df.groupby('built_in').mean()['price']
```

```python
avg_price_per_year = df.groupby('built_in').mean()['price']
avg_price_per_year.plot()
```

## histogram
to show the distribution of the prices

```python
df['price'].hist()
```

most house prices are around $200k

## barchart
avg prices per house type

```python
avg_price_per_type = df.groupby('house_type').mean()['price']
avg_price_per_type[:]
```

```python
avg_price_per_type.plot()    #default plot is line
```

```python
avg_price_per_type.plot.bar(title = 'The title')    #To make a different kind og plot    #Add a title
```

single family homes have the hightest prices

## pie chart
the number of records per house type

```python
count_per_house_type = df['house_type'].value_counts()
count_per_house_type[:]
```

```python
count_per_house_type.plot.pie()    #create a pie chart
```

## scatterplot

the relationship between price and area

```python
df.plot.scatter(x = 'area', y = 'price')
```
